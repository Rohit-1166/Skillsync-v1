## 🌐 Start Interactive Web Dashboard (Optional)
Run the cell below to spin up the premium recruiter dashboard directly on Google's cloud server. It will automatically compile the FastAPI matching server, load the indices, and generate a secure public proxy URL for you to explore the UI.

```python
# Install nest-asyncio to allow running Uvicorn inside Jupyter Notebook
!pip install nest-asyncio

import nest_asyncio
import uvicorn
import threading
import time
import socket
import sys
import os
from google.colab import output

# Ensure python knows where to import 'api' from if directory was changed
sys.path.insert(0, os.getcwd())

# Patch the event loop
nest_asyncio.apply()

# Start Uvicorn in a background thread on port 8000
# NOTE: We bind to '0.0.0.0' so Google Colab's proxy bridge can route external traffic to Uvicorn.
def run_server():
    uvicorn.run("api.app:app", host="0.0.0.0", port=8000, log_level="info")

server_thread = threading.Thread(target=run_server, daemon=True)
server_thread.start()

print("⏳ Waiting for SkillSync Matching Service to initialize and start listening on port 8000...")
start_time = time.time()
initialized = False
while time.time() - start_time < 90:
    try:
        with socket.create_connection(("127.0.0.1", 8000), timeout=1.0):
            initialized = True
            break
    except OSError:
        time.sleep(1.0)

if initialized:
    # Generate and print the secure Colab proxy URL
    proxy_url = output.eval_js("google.colab.kernel.proxyPort(8000)")
    print("\n" + "="*80)
    print("🚀 INTERACTIVE DASHBOARD IS RUNNING!")
    print(f"Click the following link to open the premium UI in your browser:")
    print(proxy_url)
    print("\nAlternatively, you can interact with the dashboard directly below:")
    print("="*80 + "\n")
    # Native Colab helper to render the dashboard inline as an iframe
    output.serve_kernel_port_as_iframe(8000, height=800)
else:
    print("❌ Error: Server failed to start or initialize within 90 seconds. Please check the logs above.")
```
