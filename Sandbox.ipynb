## 🌐 Start Interactive Web Dashboard (Optional)
Run the cell below to spin up the premium recruiter dashboard directly on Google's cloud server. It will automatically compile the FastAPI matching server, load the indices, and generate a secure public proxy URL for you to explore the UI.

```python
# Install nest-asyncio to allow running Uvicorn inside Jupyter Notebook
!pip install nest-asyncio

import nest_asyncio
import uvicorn
import threading
from google.colab import output

# Patch the event loop
nest_asyncio.apply()

# Start Uvicorn in a background thread on port 8000
def run_server():
    uvicorn.run("api.app:app", host="127.0.0.1", port=8000, log_level="info")

server_thread = threading.Thread(target=run_server, daemon=True)
server_thread.start()

# Generate and print the secure Colab proxy URL
proxy_url = output.eval_js("google.colab.kernel.proxyPort(8000)")
print("\n" + "="*80)
print("=== INTERACTIVE DASHBOARD IS RUNNING! ===")
print(f"Click the following link to open the premium UI in your browser:")
print(proxy_url)
print("="*80 + "\n")
```
